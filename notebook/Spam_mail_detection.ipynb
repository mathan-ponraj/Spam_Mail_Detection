```python
#Importing denpendencies
import numpy as np
import pandas as pd
```

```python
#Importing Logistic Regression Model
from sklearn.linear_model import LogisticRegression
```

```python
#Importing the Accuracy Score
from sklearn.metrics import accuracy_score
```

```python
#Import train test split function
from sklearn.model_selection import train_test_split
```

```python
#To convert all the strings formats into numeric
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
#Import the dataset
raw_mail_data = pd.read_csv("mail_data.csv")
raw_mail_data


```

```python
#Checking the null values
raw_mail_data.isnull().sum()

#Replacing all the null values with null string  (because the tfidfvectorizer only for strings)
mail_data =raw_mail_data.where((pd.notnull(raw_mail_data)),'')
```

```python
#Print the first five rows
mail_data.head()
```

```python
#checking the rows and columns of the dta set
mail_data.shape
```

```python
#Label spam mail as 0 and ham as 1
mail_data.loc[mail_data['Category'] == 'ham','Category'] = 1
mail_data.loc[mail_data['Category'] == 'spam','Category'] = 0
```

```python
mail_data
```

```python
#Find the totla nu of ham emai and spamemai
mail_data['Category'].value_counts()
```

```python
#Separating the data of ham email and spam email
X = mail_data['Message'] #Input feature
Y = mail_data['Category'] #Target
```

```python
X
```

```python
Y
```

```python
#Splitting the data into training data and testing data
#Training data = X_tain,Y_train
#Testing data = X_test,Y_test
X_train,X_test,Y_train,Y_test = train_test_split(X,Y, test_size = 0.2, random_state = 2)
```

```python
#Tranforming the text data   feature vectors that can be uesd as input to the logistic regression
feature_extraction = TfidfVectorizer(min_df = 1,stop_words = "english", lowercase = True)

# min_df = minimum document frequency
#stop_words = eglish words
```

```python
X_train_num = feature_extraction.fit_transform(X_train)
X_test_num = feature_extraction.transform(X_test)
```

```python
#Convert Y_train and Y_test values as integer
Y_train = Y_train.astype('int')
Y_test = Y_test.astype('int')
```

```python
#@title Training model
```

```python
#@title Logistic Regression
model = LogisticRegression()

```

```python
#Training the logistic Regression model with training data
model.fit(X_train_num,Y_train)

```

```python
#@title Evaluating the Training model
X_train_prediction = model.predict(X_train_num)
```

```python
#Accuracy score for Training data
training_data_accuracy = accuracy_score(Y_train,X_train_prediction)
print("Accuracy score of Training data is",training_data_accuracy)
```

```python
# Prediction on test data
#@title Evaluating the Testing model
X_test_prediction = model.predict(X_test_num)
#Accuracy score for Testing data
testing_data_accuracy = accuracy_score(Y_test,X_test_prediction)
print("Accuracy score of Testing data is",testing_data_accuracy)
```

```python
#@title Build the predictive system
input_mail = ["hello sir,your'e selected for free trail of car and if you are lucky now you can win a car free free free"]

#convert text to feature vectors
input_data_features = feature_extraction.transform(input_mail)

#making prediction
prediction = model.predict(input_data_features)
print(prediction)
if (prediction == 0):
  print("The email is spam")
else:
  print("The email is ham")
```
